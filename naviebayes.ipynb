```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
news_data=fetch_20newsgroups(subset='all')
```

```python
news_train=fetch_20newsgroups(subset='train')
```

```python
news_test=fetch_20newsgroups(subset='test')
```

```python
print("all target labels",news_train.target_names)
```

```python
categories=['alt.atheism','comp.graphics','sci.space']
```

```python
news_train_subset=fetch_20newsgroups(subset='train',categories=categories)
```

```python
news_test_subset=fetch_20newsgroups(subset='test',categories=categories)
```

```python
print("new training set target names:",news_train_subset.target_names)
```

```python
print("5 th article in training data\n",news_train_subset.data[4]) 
```

```python
print("shape of trainin gdata ",len(news_train_subset.data))
```

```python
print("shape of training targets:",len(news_train_subset.target))
```

```python
print("training set filenames\n",news_train_subset.target[:5])
```

```python
vectorizer=CountVectorizer()
X_train_counts=vectorizer.fit_transform(news_train_subset.data)
```

```python
clf=BernoulliNB()
clf.fit(X_train_counts,news_train_subset.target)
```

```python
X_test_counts=vectorizer.transform(news_test_subset.data)
```

```python
y_pred=clf.predict(X_test_counts)
```

```python
accuracy=accuracy_score(news_test_subset.target,y_pred)
print("Accuracy:",accuracy)
```

```python
tfidf_vectorizer=TfidfVectorizer()
X_train_tfidf=tfidf_vectorizer.fit_transform(news_train_subset.data)
clf_tfidf=MultinomialNB()
clf_tfidf.fit(X_train_tfidf,news_train_subset.target)
# Transform test data
X_test_tfidf = tfidf_vectorizer.transform(news_test_subset.data)
y_pred_tfidf = clf_tfidf.predict(X_test_tfidf)
```

```python
accuracy_tfidf = accuracy_score(news_test_subset.target, y_pred_tfidf)
print("Test set accuracy with TfidfVectorizer:", accuracy_tfidf)

```
