## <span style="color:darkred; font-size: 24px;">üìå What is Classification?</span>

Classification is a **supervised learning** technique in machine learning and statistics where the goal is to **categorize data into predefined classes**.

üîπ It determines **which category** a new observation belongs to based on past data.  
üîπ The model is trained with **labeled data** (data where the correct category is already known).  
üîπ Examples:
   - Identifying if an **email** is **spam or not** üìß  
   - Classifying **images** as **cat üê± or dog üê∂**  
   - Predicting whether a **tumor is malignant or benign** based on medical data.

---

## <span style="color:darkblue; font-size: 24px;">üìñ What is K-Nearest Neighbors (KNN)?</span>

üîπ **K-Nearest Neighbors (KNN)** is a **simple yet powerful** classification algorithm.  
üîπ It works by finding the **K** closest training data points (neighbors) to a new data point and **assigning it the most common class**.  
üîπ KNN is a **non-parametric** and **lazy learning algorithm** (it does not learn an explicit model).

### <span style="color:green; font-size: 22px;">üõ† How Does KNN Work?</span>

1Ô∏è‚É£ Choose the number of **neighbors (K)**.  
2Ô∏è‚É£ Calculate the **distance** between the new point and existing points.  
3Ô∏è‚É£ Identify the **K-nearest neighbors**.  
4Ô∏è‚É£ Assign the class label based on a **majority vote** among the neighbors.

## <span style="color:purple; font-size: 24px;">üìè Common Distance Metrics</span>

KNN uses different **distance metrics** to measure how close two points are.

üîπ **Euclidean Distance** (Most common)  
   $$
   d(p, q) = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2}
   $$

üîπ **Manhattan Distance**  
   $$
   d(p, q) = \sum_{i=1}^{n} |q_i - p_i|
   $$

üîπ **Minkowski Distance** (Generalized form of Euclidean and Manhattan)  
   $$
   d(p, q) = \left( \sum_{i=1}^{n} |q_i - p_i|^p \right)^{\frac{1}{p}}
   $$

## <span style="color:purple; font-size: 24px;">üìè Numerical Examples for Distance Metrics</span>

Let's consider two points:  
**P (2, 3)** and **Q (5, 7)**  

---

### <span style="color:darkblue; font-size: 22px;">üîπ Euclidean Distance</span>

$$
d(P, Q) = \sqrt{(5 - 2)^2 + (7 - 3)^2}
$$

$$
= \sqrt{3^2 + 4^2}
$$

$$
= \sqrt{9 + 16} = \sqrt{25} = 5
$$

üìå **Euclidean Distance = 5**

---

### <span style="color:darkred; font-size: 22px;">üîπ Manhattan Distance</span>

$$
d(P, Q) = |5 - 2| + |7 - 3|
$$

$$
= 3 + 4 = 7
$$

üìå **Manhattan Distance = 7**

---

### <span style="color:darkgreen; font-size: 22px;">üîπ Minkowski Distance (p = 3)</span>

$$
d(P, Q) = \left( |5 - 2|^3 + |7 - 3|^3 \right)^{\frac{1}{3}}
$$

$$
= \left( 3^3 + 4^3 \right)^{\frac{1}{3}}
$$

$$
= \left( 27 + 64 \right)^{\frac{1}{3}}
$$

$$
= (91)^{\frac{1}{3}} \approx 4.49
$$

üìå **Minkowski Distance (p=3) ‚âà 4.49**


```python
from sklearn.metrics.pairwise import euclidean_distances, manhattan_distances
from sklearn.metrics.pairwise import pairwise_distances

# Define two points in 3D space
P = [[2, 3, 5]]
Q = [[5, 7, 9]]

# Euclidean Distance
euclidean_dist = euclidean_distances(P, Q)[0][0]

# Manhattan Distance
manhattan_dist = manhattan_distances(P, Q)[0][0]

# Minkowski Distance (p=3)
minkowski_dist = pairwise_distances(P, Q, metric='minkowski', p=3)[0][0]

# Print results
print(f"Euclidean Distance: {euclidean_dist:.2f}")
print(f"Manhattan Distance: {manhattan_dist:.2f}")
print(f"Minkowski Distance (p=3): {minkowski_dist:.2f}")
```

# <span style="color:blue; font-size: 30px;">Choosing the Right K in KNN</span>

---

## <span style="color:darkred; font-size: 24px;">üìå Why is Choosing K Important?</span>

The value of **K (number of neighbors)** in KNN affects the **accuracy and generalization** of the model.

- **Too small K** (e.g., 1 or 3): The model becomes sensitive to noise and may **overfit**.
- **Too large K** (e.g., 15, 20): The model smooths too much and may **underfit**.
- **Optimal K**: Found by trying different values and selecting the best based on **validation accuracy**.

---

## <span style="color:darkblue; font-size: 24px;">üìä Finding the Best K Using Cricket Dataset</span>

We use **training data** to fit the model and **test data** to evaluate the accuracy for different values of **K**.

---

## <span style="color:purple; font-size: 24px;">üìå Steps for Selecting K</span>

1Ô∏è‚É£ **Train KNN on different values of K (1 to 15)**  
2Ô∏è‚É£ **Measure Accuracy for each K**  
3Ô∏è‚É£ **Plot Accuracy vs. K** to find the optimal value  

---


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define dataset (Matches, Runs, Wickets)
X_train = np.array([
    [463, 18426, 154], [103, 476, 114], [111, 675, 150], [304, 8701, 110],
    [120, 1544, 173], [147, 1990, 171], [225, 3783, 253], [229, 883, 315],
    [45, 731, 44], [308, 11363, 100], [271, 938, 334], [222, 10533, 4]
])

y_train = np.array(["Batsman", "Bowler", "Bowler", "All Rounder", "All Rounder",
                    "All Rounder", "All Rounder", "Bowler", "All Rounder",
                    "Batsman", "Bowler", "Batsman"])

X_test = np.array([
    [201, 7808, 8], [251, 8273, 94], [226, 5615, 36], [234, 1237, 265], [200, 792, 269]
])

y_test = np.array(["Batsman", "Batsman", "Batsman", "Bowler", "Bowler"])

# Ensure k is not greater than number of training samples
max_k = len(X_train)  # Max k should be ‚â§ number of training samples

# Try different values of K
k_values = range(1, max_k + 1)
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k, metric="euclidean")
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(k, acc)

# Plot Accuracy vs. K
plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker="o", linestyle="-", color="blue")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy")
plt.title("Choosing the Best K for KNN")
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Best K selection
best_k = k_values[np.argmax(accuracies)]
print(f"Best value of K: {best_k}")
```

# <span style="color:blue; font-size: 30px;">üîç Grid Search for Optimizing K in KNN</span>

---

## <span style="color:darkred; font-size: 24px;">üìå Why Use Grid Search?</span>

- Instead of manually trying different values of **K**, Grid Search **automatically finds the best K**.
- It searches for **optimal hyperparameters** by evaluating different combinations.
- It can also tune **distance metrics** (e.g., Euclidean, Manhattan).

---

## <span style="color:darkblue; font-size: 24px;">üõ† Steps in Grid Search</span>

1Ô∏è‚É£ **Define possible values for K** (e.g., 1 to 15).  
2Ô∏è‚É£ **Choose distance metrics** (Euclidean, Manhattan, etc.).  
3Ô∏è‚É£ **Perform cross-validation** to evaluate each combination.  
4Ô∏è‚É£ **Select the best K based on accuracy**.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Define KNN model
knn = KNeighborsClassifier()

# Define hyperparameters to tune
param_grid = {'n_neighbors': range(1, 5), 'metric': ['euclidean', 'manhattan']}

# Perform Grid Search
grid_search = GridSearchCV(knn, param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Best parameters and accuracy
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Accuracy: {grid_search.best_score_:.2f}")
```

# <span style="color:blue; font-size: 30px;">üîÑ Cross-Validation in Machine Learning</span>

---

## <span style="color:darkred; font-size: 24px;">üìå What is Cross-Validation?</span>

- **Cross-validation (CV)** is a technique to **evaluate the performance** of a machine learning model.
- Instead of using a **single train-test split**, CV splits the data into **multiple folds** to test the model on different subsets.
- It ensures the model **generalizes well** to unseen data.

---

## <span style="color:darkblue; font-size: 24px;">üîπ Why Use Cross-Validation?</span>

‚úÖ Prevents **overfitting** by testing the model on multiple subsets.  
‚úÖ Ensures the model performs well on **unseen data**.  
‚úÖ Provides a **more reliable estimate** of model performance.  

---

## <span style="color:purple; font-size: 24px;">üìä Types of Cross-Validation</span>

### **1Ô∏è‚É£ K-Fold Cross-Validation**
- The dataset is split into **K folds** (e.g., 5).
- The model is trained on **K-1 folds** and tested on **the remaining fold**.
- The process is repeated **K times**, and the average score is computed.

### **2Ô∏è‚É£ Stratified K-Fold Cross-Validation**
- Similar to K-Fold but ensures **each fold has an equal distribution of classes**.
- Useful for **imbalanced datasets**.

### **3Ô∏è‚É£ Leave-One-Out Cross-Validation (LOO)**
- Each sample is **treated as a separate test set** while the rest are used for training.
- Good for **small datasets**, but computationally expensive.

# <span style="color:blue; font-size: 30px;">Curse of Dimensionality in KNN</span>

---

## <span style="color:darkred; font-size: 24px;">üìå Curse of Dimensionality</span>

### **üîπ Why Does KNN Perform Poorly in High Dimensions?**
- As the **number of features (dimensions) increases**, KNN struggles because:
  - üîπ **Distances become less meaningful** ‚Üí All points seem equally distant.
  - üîπ **Data sparsity increases** ‚Üí Points are spread far apart in a large space.
  - üîπ **Computational cost rises** ‚Üí Distance calculations become expensive.

### **üîπ Solution: Dimensionality Reduction**
1Ô∏è‚É£ **Use PCA (Principal Component Analysis)** ‚Üí Reduces dimensions while preserving variance.  
2Ô∏è‚É£ **Use LDA (Linear Discriminant Analysis)** ‚Üí Maximizes class separability.  
3Ô∏è‚É£ **Select important features** ‚Üí Use feature selection techniques.
